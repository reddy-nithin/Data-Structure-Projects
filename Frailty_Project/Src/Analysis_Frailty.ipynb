# Importing Required Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Analysis

```python
# reading clean_data_frailty
df = pd.read_csv('https://raw.githubusercontent.com/reddy-nithin/PDS-Assignments/refs/heads/main/Assignment%201/Frailty_Project/Clean_Data/clean_frailty_data.csv')
```

```python
df
```

```python
# Generating heat map to see the correlation.
sns.heatmap(df.corr(), annot=True)
df.corr()
```

```python
# calculating BMI from the data set
df['BMI'] = df['Weight_kg'] / ((df['Height_cms']/100)**2)
df
sns.heatmap(df.corr(), annot=True)
df.corr()
```

```python
# plotting bmi vs grip strength
sns.scatterplot(x='BMI', y='Grip strength', data=df)
```

```python
#ploting BMI and Frailty
sns.scatterplot(x='BMI', y='Frailty', data=df)
```

```python
# plotting grip strength vs failty
sns.scatterplot(x='Grip strength', y='Frailty', data=df)
```

```python
# Bar chart for Frailty vs Grip Strength
plt.figure(figsize=(8, 6))
sns.barplot(x='Grip strength', y='Frailty', data=df)
plt.xlabel('Grip Strength')
plt.ylabel('Frailty')
plt.title('Frailty vs Grip Strength')
plt.show()
```

```python
# Line graph for Frailty vs Grip Strength
plt.figure(figsize=(8, 6))
sns.lineplot(x='Grip strength', y='Frailty', data=df)
plt.xlabel('Grip Strength')
plt.ylabel('Frailty')
plt.title('Frailty vs Grip Strength')
plt.show()
```
