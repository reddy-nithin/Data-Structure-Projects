```python
#importing required packaged
import pandas as pd
import numpy as np
```

```python
# Reading the data file
df = pd.read_csv('https://raw.githubusercontent.com/reddy-nithin/PDS-Assignments/refs/heads/main/Assignment%201/StudentPerformance_Project/Raw_data/StudentsPerformance.csv')
df.head()
```

```python
# Checking for null values
df.info()

df.isnull().sum()
```

```python
# Printing only parental level of education values
df['parental level of education'].value_counts()
```

```python
# Checking for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")

# Checking for inconsistent data in categorical columns (e.g., "gender", "race/ethnicity", etc.)
for column in df.select_dtypes(include=['object']):
  print(f"\nUnique values for {column}:")
  print(df[column].unique())

print(df.head())
```

```python
# writing the new cleaned data file
df.to_csv('cleaned_data_studentPerformance.csv', index=False)
```
