# Importing required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Data Analysis

```python
df = pd.read_csv('https://raw.githubusercontent.com/reddy-nithin/PDS-Assignments/refs/heads/main/Assignment%201/StudentPerformance_Project/Raw_data/StudentsPerformance.csv')
df
```

```python
# adding a new column "Total Score"
df['Total Score'] = df['math score'] + df['reading score'] + df['writing score']
df
```

```python
# 1. Bar Chart: Count of Students by Race/Ethnicity
fig1 = px.histogram(df, x="race/ethnicity",
                   title="Count of Students by Race/Ethnicity")
fig1.show()
```

```python
# 2. Scatter Plot: Math vs. Reading Scores
fig2 = px.scatter(df, x="math score", y="reading score",
                 color="gender",
                 title="Math vs. Reading Scores by Gender")
fig2.show()
```

```python
# 3. Math Score Distribution by Gender
fig3 = px.histogram(df, x="math score", color="gender", nbins=20, title="Math Score Distribution by Gender")
fig3.show()
```

```python
# 4. Total Score by Parental Education Level
fig4 = px.bar(df.groupby('parental level of education')['Total Score'].mean().reset_index(),
            x='parental level of education',
            y='Total Score',
            title='Average Total Score by Parental Education Level')
fig4.show()
```

```python
# 5. Bar Chart of Average Total Score by Test Preparation Course
fig5 = px.bar(df.groupby('test preparation course')['Total Score'].mean().reset_index(), x='test preparation course', y='Total Score', title='Average Total Score by Test Preparation Course')
fig5.show()
```

```python
# 6. Box Plot of Total Scores by Lunch
fig6 = px.box(df, x='lunch', y='Total Score', title='Total Scores by Lunch')
fig6.show()
```

```python
# 7. Line Plot of Average Total Score by Gender and Race/Ethnicity
fig7 = px.line(df.groupby(['gender', 'race/ethnicity'])['Total Score'].mean().reset_index(), x='race/ethnicity', y='Total Score', color='gender', title='Average Total Score by Gender and Race/Ethnicity')
fig7.show()
```

```python
# 8. Pie Chart of Percentage of Students in Different Total Score Ranges
bins = [0, 150, 200, 250, 300]
labels = ['0-150', '151-200', '201-250', '251-300']
df['Total Score Range'] = pd.cut(df['Total Score'], bins=bins, labels=labels)
fig8 = px.pie(df, names='Total Score Range', title='Percentage of Students in Different Total Score Ranges')
fig8.show()
```

```python
# 9. Scatter plot of reading vs writing scores, colored by parental education level
fig9 = px.scatter(df, x="reading score", y="writing score", color="parental level of education",
                  title="Reading vs. Writing Scores by Parental Education")
fig9.show()
```
