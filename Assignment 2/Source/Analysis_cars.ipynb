```python
import numpy as np
import pandas as pd
```

# Data Analysis

```python
 url = "https://raw.githubusercontent.com/reddy-nithin/PDS-Assignments/refs/heads/main/Assignment%202/Clean%20Data/clean_data_cars.csv"
 df = pd.read_csv(url)
 df.head()
```

### Adding more Features

```python
from datetime import datetime

current_year = datetime.now().year
df['Car_Age'] = current_year - df['Year']
df.head()
```

```python
usage_threshold = 50000
df['Usage_Category'] = df['Kilometers_Driven'].apply(
    lambda km: 'Fairly Used' if km < usage_threshold else 'Used More')
df.head()
```

### Performing select, filter, rename, mutate, arrange and summarize with group by operations  on this dataset.

```python
# Using Select and Filter
selected_data = df[['Name', 'Location', 'Car_Age', 'Mileage', 'Price', 'Usage_Category']]
filtered_data = selected_data[(selected_data['Car_Age'] <= 5) & (selected_data['Price'] < 1000000) & (selected_data['Usage_Category'] == 'Fairly Used')]
filtered_data
```

```python
# Renaming columns for better clarity
filtered_data = filtered_data.rename(columns={'Name': 'Car_Model', 'Price': 'Price_when_new(Lakhs)'})
filtered_data.head()
```

```python
# Mutating the data with new "Car_Make" Column
filtered_data['Car_Make'] = filtered_data['Car_Model'].str.split().str[0]
filtered_data.Car_Make.head()
```

```python
# Arranging data by 'Selling_Price' in ascending order
arranged_data = filtered_data.sort_values(by='Price_when_new(Lakhs)')
arranged_data.head()
```

```python
# Summarizing with Group By: Calculating the average price by 'Location' for cars
summary = arranged_data.groupby('Location').agg(Average_Selling_Price=('Price_when_new(Lakhs)', 'mean')).reset_index()
summary.head()
```
